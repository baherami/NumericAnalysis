```python
%pylab inline
from math import factorial
```

```python
def approximateSin(x,n):
    sign=1
    i=1
    statement=0
    power=1
    while(i<=n):
        statement += sign * x ** i / factorial(i)
        i += 2
        sign = sign * -1
    return statement

        
        
```
