# 1.a

```python
def approx_log(x,iterat):
    n=1
    sumx=0
    sign=1.0
    while(n<(iterat+1)):
        sumx=sumx+ (sign* (1/n)*((x-1)**n) )
        sign=sign * (-1)
        n+=1
    return sumx
x=arange(-5,5,0.01)
y=approx_log(x,5)

figure()
plot(x,y)
```

# 1.b

```python
x=2.0
n=5
true_value =log(x)
approximation= approx_log(x,n)
Et=true_value -approximation
relative_true_error=Et/true_value
print("aprox=",approximation,"true value",true_value)

print("true error is",Et)
print("relative true error is",relative_true_error)
```

# 1.c

```python
def approx_log2(x,es):
    n=1
    sumx=0.0
    appx=0.0
    ea=1
    prevsumx=0.0
    sign=1.0
    while True:
        appx=(sign* (1/n)*((x-1)**n))
        sign=sign * -1
        prevsumx=sumx
        sumx=prevsumx+appx 
        ea=abs((sumx - prevsumx)/sumx) *100
        n=n+1
        if(n>100000) or (ea<=es):
            break
    print("ea=",ea,"es=",es,"log(2.0)=",log(x),"approximation of log(2.0)=",sumx)
    return n

n=3
es=(0.5*10**(2-n))
x=2.0
print("number of terms needed,",approx_log2(x,es))
```
