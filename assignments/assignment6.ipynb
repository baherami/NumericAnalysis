## 1. a
Consider $T_1$, $T_2$, $T_3$ and  $a$ as the variables. There are four equations and four variables so the problem can have a solution.
The variable matrix contains a last element which is $a$.


```python
g = 9.81
theta = 45.0
m=array( [10,4,5,6.0])
u=array([0.25,0.30,0.20,1])
b=sin(theta)- cos(theta)*u
b[3]=-1.0
b=b*g*m
A=array([[1,0,0,m[0]],
         [-1,1,0,m[1]],
         [0,-1,1,m[2]],
         [0,0,-1,m[3]],
        ])
T = solve(A, b)
print("T1={0:.2f},T2={1:.2f},T3={2:.2f},a={3:.2f}".format(T[0],T[1],T[2],T[3]))
```
